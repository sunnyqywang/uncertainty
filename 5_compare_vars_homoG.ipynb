```python
import numpy as np
import pandas as pd
from setup import *
```

```python
result_file = "results.csv"
```

```python
group_cols = ['Model Type','Distribution','homo']
metric_cols = ['train_nll_loss','val_nll_loss','test_nll_loss',
                 'train_mae','val_mae','test_mae',
                 'train_mpiw','val_mpiw','test_mpiw',
                 'train_picp','val_picp','test_picp']
table_cols = ['Model', 'Lookback','test_nll_loss',
                 'test_mpiw',
                 'test_picp',
                 'test_mae',
                 'test_mape']
```

```python
b_test_mean = 635.398
a_test_mean = 88.963
```

## Before

```python
df1 = pd.read_csv(project_dir+"results/220430_before/"+result_file)
df2 = pd.read_csv(project_dir+"results/220517_before/"+result_file)
df3 = pd.read_csv(project_dir+"results/220518_before/"+result_file)
df4 = pd.read_csv(project_dir+"results/220519_before/"+result_file)

df1['homo'] = 0.5
df2['homo'] = 1
df3['homo'] = 0.25
df4['homo'] = 0.2

dfb = pd.concat([df1, df2, df3, df4]).reset_index()
# df1 = df1[df1['Model Type'] != 'GAT']
# df1 = process_results_df(df1)
dfb['test_mape'] = dfb['test_mae'] / b_test_mean
```

```python
# top 5 by val nll loss 
dfb[((dfb['Model Type']=='GAT')&(dfb['Lookback']==6)&(dfb['Distribution']=='norm_homo'))].sort_values(by='val_nll_loss').head(6)[group_cols+table_cols]
```

```python
dfb[((dfb['Model Type']=='GCN')&(dfb['Lookback']==6)&(dfb['Distribution']=='norm_homo'))].sort_values(by='val_nll_loss').head(6)[group_cols+table_cols]
```

```python
# top model by category
dfb = dfb.sort_values(by='val_nll_loss')
dfb.groupby(group_cols).head(2)[group_cols+table_cols].sort_values(group_cols)


# dfb.loc[dfb.groupby(group_cols)['val_nll_loss'].idxmin()][group_cols+table_cols].sort_values(group_cols)
```

```python
dfb.groupby(['Model Type','Distribution','Lookback','homo']).count()['Period']
```

```python
# median in each category
dfb.groupby(['Model Type','Distribution','Lookback']).median()[[
                                     'train_nll_loss','val_nll_loss','test_nll_loss',
                                     'train_mae','val_mae','test_mae',
                                     'train_mpiw','val_mpiw','test_mpiw',
                                     'train_picp','val_picp','test_picp']].round(3)
```

## After

```python
df1 = pd.read_csv(project_dir+"results/220430_after/"+result_file)
df2 = pd.read_csv(project_dir+"results/220517_after/"+result_file)
df3 = pd.read_csv(project_dir+"results/220518_after/"+result_file)
df4 = pd.read_csv(project_dir+"results/220519_after/"+result_file)
df5 = pd.read_csv(project_dir+"results/220520_after/"+result_file)

df1['homo'] = 0.5
df2['homo'] = 1
df3['homo'] = 0.25
df4['homo'] = 0.4
df5['homo'] = 0.45

dfa = pd.concat([df1, df2, df3, df4, df5]).reset_index()

# df1 = df1[df1['Model Type'] != 'GAT']
# df1 = process_results_df(df1)
dfa['test_mape'] = dfa['test_mae'] / a_test_mean
```

```python
# top 5 by val nll loss (for ensembling)
dfa[((dfa['Model Type']=='GAT')&(dfa['Lookback']==6)&(dfa['Distribution']=='norm_homo'))].sort_values(by='val_nll_loss').head(6)[group_cols+table_cols]
```

```python
dfa[((dfa['Model Type']=='GCN')&(dfa['Lookback']==6)&(dfa['Distribution']=='norm_homo'))].sort_values(by='val_nll_loss').head(6)[group_cols+table_cols]
```

```python
# top models by model category
dfa = dfa.sort_values(by='val_nll_loss')
dfa.groupby(group_cols).head(2)[group_cols+['val_nll_loss']+table_cols].sort_values(group_cols)
```
