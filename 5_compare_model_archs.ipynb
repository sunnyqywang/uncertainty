```python
import numpy as np
import pandas as pd
from setup import *
```

```python
out_folder_1 = "2104_rc_tnorm/"
out_folder_2 = "2104_rc_new1/"

result_file = "rail_catchment_mve_results.csv"
```

```python
def process_results_df(results):
    results['train_rmse'] = np.sqrt(results['train_mse'])
    results['test_rmse'] = np.sqrt(results['test_mse'])
    results['Adjacency'] = pd.Categorical(results['Adjacency'], ['func','euc','con','net','func_euc','func_con','func_net','func_euc_con_net'])
    results = results[results['test_nll_loss']<1000]

    return results
```

```python
df1 = pd.read_csv(project_dir+"results/"+out_folder_1+result_file)
df1 = process_results_df(df1)

df2 = pd.read_csv(project_dir+"results/"+out_folder_2+result_file)
df2 = process_results_df(df2)
```

```python
df1.groupby(['Period','Adjacency']).mean()[['train_nll_loss','test_nll_loss','test_u']].round(2)
```
