```python
import numpy as np
import pandas as pd
from setup import *
```

```python
result_file = "results.csv"
```

```python
group_cols = ['Model Type','Distribution']
metric_cols = ['train_nll_loss','val_nll_loss','test_nll_loss',
                 'train_mae','val_mae','test_mae',
                 'train_mpiw','val_mpiw','test_mpiw',
                 'train_picp','val_picp','test_picp']
table_cols = ['Model','Lookback','test_nll_loss',
                 'test_mpiw',
                 'test_picp',
                 'test_mae',
                 'test_mape']
```

```python
b_test_mean = 635.398
a_test_mean = 88.963
```

## Before

```python
df1 = pd.read_csv(project_dir+"results/220405_before/"+result_file)
df2 = pd.read_csv(project_dir+"results/220421_before/"+result_file)

dfb = pd.concat([df1, df2]).reset_index()
# df1 = df1[df1['Model Type'] != 'GAT']
# df1 = process_results_df(df1)
dfb['test_mape'] = dfb['test_mae'] / b_test_mean
```

```python
# top 5 by val nll loss (for ensembling)
print(dfb[((dfb['Model Type']=='GAT')&(dfb['Lookback']==4)&(dfb['Distribution']=='norm'))].sort_values(by='val_nll_loss').head(6)[['Model Type','Model']])
print(dfb[((dfb['Model Type']=='GCN')&(dfb['Lookback']==4)&(dfb['Distribution']=='norm'))].sort_values(by='val_nll_loss').head(6)[['Model Type','Model']])
```

```python
# top model by category
dfb.loc[dfb.groupby(group_cols)['val_nll_loss'].idxmin()][group_cols+table_cols].sort_values(group_cols)
```

```python
dfb.groupby(['Model Type','Distribution','Lookback']).count()['Period']
```

```python
# median in each category
dfb.groupby(['Model Type','Distribution','Lookback']).median()[[
                                     'train_nll_loss','val_nll_loss','test_nll_loss',
                                     'train_mae','val_mae','test_mae',
                                     'train_mpiw','val_mpiw','test_mpiw',
                                     'train_picp','val_picp','test_picp']].round(3)
```

## After

```python
df1 = pd.read_csv(project_dir+"results/220405_after/"+result_file)
df2 = pd.read_csv(project_dir+"results/220421_after/"+result_file)

dfa = pd.concat([df1, df2]).reset_index()
# df1 = df1[df1['Model Type'] != 'GAT']
# df1 = process_results_df(df1)
dfa['test_mape'] = dfa['test_mae'] / a_test_mean
```

```python
# top 5 by val nll loss (for ensembling)
print(dfa[((dfa['Model Type']=='GAT')&(dfa['Lookback']==4)&(dfa['Distribution']=='norm'))].sort_values(by='val_nll_loss').head(6)[['Model Type','Model']])
print(dfa[((dfa['Model Type']=='GCN')&(dfa['Lookback']==4)&(dfa['Distribution']=='norm'))].sort_values(by='val_nll_loss').head(6)[['Model Type','Model']])
```

```python
# top models by model category
dfa = dfa.sort_values(by='val_nll_loss')
dfa.groupby(group_cols).head(2)#[group_cols+table_cols].sort_values(group_cols)
```
