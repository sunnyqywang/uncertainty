```python
import numpy as np
import pandas as pd
from setup import *
```

```python
result_file = "results.csv"
```

```python
group_cols = ['Model Type','Distribution','Time Size']
metric_cols = ['train_nll_loss','test_nll_loss',
                 'train_mae','test_mae',
                 'train_mape','test_mape',
                 'train_mpiw','test_mpiw',
                 'train_picp','test_picp']
table_cols = ['Model','weight_decay','dropout','val_nll_loss','test_nll_loss',
                 'test_mpiw',
                 'test_picp',
                 'test_mae',
                 'test_mape']
```

## Before

```python
dfb = pd.read_csv(project_dir+"results/R1_rail_190801-200301_200302-200315/"+result_file)
```

```python
# top model by category
dfb = dfb.sort_values(by='val_nll_loss')
dfb.groupby(group_cols).head(1)[group_cols+table_cols+['n_head']].sort_values(group_cols)


# dfb.loc[dfb.groupby(group_cols)['val_nll_loss'].idxmin()][group_cols+table_cols].sort_values(group_cols)
```

```python
dfb[((dfb['Model Type']=='GCN')&(dfb['Distribution']=='norm'))].sort_values(by='val_nll_loss').head()[group_cols+table_cols]
```

```python
# top 5 by val nll loss (for ensembling)
dfb[((dfb['Model Type']=='GAT')&(dfb['Distribution']=='norm'))].sort_values(by='val_nll_loss').head(5)[group_cols+table_cols]
```

```python
dfb.groupby(['Model Type','Distribution','Time Size']).count()['Lookback']
```
